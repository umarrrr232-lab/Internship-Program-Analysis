# Internship Completion Rate Analysis
**internee.pk Task**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
print(df.shape)
df.head()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# drop the empty column
df = df.drop(columns=['Unnamed: 16'])

# drop rows where status is missing
df = df.dropna(subset=['Status Description'])

print('Cleaned! Rows remaining:', len(df))
```

```python
# see all status values
df['Status Description'].value_counts()
```

```python
# group statuses into simpler categories
def categorize(status):
    if status in ['Team Allocated', 'Rewards Award']:
        return 'Completed'
    elif status in ['Dropped Out', 'Withdraw']:
        return 'Dropped'
    elif status == 'Rejected':
        return 'Rejected'
    else:
        return 'Enrolled'

df['Category'] = df['Status Description'].apply(categorize)
df['Category'].value_counts()
```

```python
# summary
total    = len(df)
comp     = (df['Category'] == 'Completed').sum()
dropped  = (df['Category'] == 'Dropped').sum()
enrolled = (df['Category'] == 'Enrolled').sum()

print('Total Learners  :', total)
print('Completed       :', comp,  f'({comp/total*100:.1f}%)')
print('Dropped Out     :', dropped, f'({dropped/total*100:.1f}%)')
print('Still Enrolled  :', enrolled, f'({enrolled/total*100:.1f}%)')
```

```python
# pie chart of categories
counts = df['Category'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Internship Completion Overview')
plt.show()
```

```python
# bar chart - outcomes by opportunity category
grouped = df.groupby(['Opportunity Category', 'Category']).size().unstack(fill_value=0)

grouped.plot(kind='bar', figsize=(9,5))
plt.title('Outcomes by Opportunity Category')
plt.xlabel('Category')
plt.ylabel('Number of Learners')
plt.xticks(rotation=30)
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()
```

```python
# dropout rate by major (top 8 majors only)
major_counts = df['Current/Intended Major'].value_counts().head(8).index
major_df = df[df['Current/Intended Major'].isin(major_counts)]

major_group = major_df.groupby(['Current/Intended Major', 'Category']).size().unstack(fill_value=0)
major_group['Dropout Rate'] = major_group.get('Dropped', 0) / major_group.sum(axis=1) * 100

major_group['Dropout Rate'].sort_values().plot(kind='barh', figsize=(9,5), color='salmon')
plt.title('Dropout Rate by Major (Top 8)')
plt.xlabel('Dropout Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# top 5 countries by enrollment
top_countries = df['Country'].value_counts().head(5)

top_countries.plot(kind='bar', figsize=(8,4), color='steelblue')
plt.title('Top 5 Countries by Enrollment')
plt.xlabel('Country')
plt.ylabel('Number of Learners')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# gender distribution
gender = df['Gender'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', startangle=90)
plt.title('Enrollment by Gender')
plt.show()
```
